phase 2

```python
def random_function(x):
    result = np.zeros_like(x)
    
    result[x < -5] = (x[x < -5])
    result[(x >= -5) & (x <= 5)] = np.arctan(x[(x >= -5) & (x <= 5)])
    result[x > 5] = x[x > 5]*(3/2)
    
    return result
    
x = np.linspace(-10, 10, 1000)
y = random_function(x)
g = Learn()
print(g.fit(x, y).traverse())
```
